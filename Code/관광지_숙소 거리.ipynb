```python
import pandas as pd
```

```python
df=pd.read_excel('천안시 관광지_콘텐츠 추천.xlsx')
df1=pd.read_excel('천안시 숙소_콘텐츠 추천.xlsx')
df
```

```python
df1
```

```python
df = df.rename(columns={
    '도로명주소': '관광지주소',
    'Latitude': '관광지 Latitude',
    'Longitude': '관광지 Longitude',
    '중분류 카테고리': '관광지 중분류',
    '소분류 카테고리': '관광지 소분류',
    '검색건수': '관광지 검색건수',
    '유사도1' : '관광지 유사도1',
    '유사도2' : '관광지 유사도2',
    '유사도3' : '관광지 유사도3'
})

df = df[['관광지명', '관광지주소', '관광지 Latitude', '관광지 Longitude', '관광지 중분류', '관광지 소분류', '관광지 검색건수','유사 관광지명','관광지 유사도1','관광지 유사도2','관광지 유사도3']]
df
```

```python
df1 = df1.rename(columns={
    '관광지명': '숙소명',
    '도로명주소': '숙소주소',
    'Latitude': '숙소 Latitude',
    'Longitude': '숙소 Longitude',
    '중분류 카테고리': '숙소 중분류',
    '소분류 카테고리': '숙소 소분류',
    '검색건수': '숙소 검색건수',
    '유사 관광지명' : '유사 숙소명',
    '유사도1' : '숙소 유사도1',
    '유사도2' : '숙소 유사도2',
    '유사도3' : '숙소 유사도3'
})

df1 = df1[['숙소명', '숙소주소', '숙소 Latitude', '숙소 Longitude', '숙소 중분류', '숙소 소분류', '숙소 검색건수', '반려동물 가능여부', '유사 숙소명', '숙소 유사도1','숙소 유사도2','숙소 유사도3']]
df1
```

```python
from geopy.distance import geodesic

def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).km

merged_df = []

for i, row in df.iterrows():
    관광지_lat = row['관광지 Latitude']
    관광지_lon = row['관광지 Longitude']
    
    df1['거리'] = df1.apply(lambda x: calculate_distance(관광지_lat, 관광지_lon, x['숙소 Latitude'], x['숙소 Longitude']), axis=1)
    
    nearby_attractions = df1[df1['거리'] <= 2].copy()
    
    if not nearby_attractions.empty:
        for _, attraction_row in nearby_attractions.iterrows():
            combined_row = row.to_dict()
            combined_row.update(attraction_row.to_dict())
            merged_df.append(combined_row)

final_df = pd.DataFrame(merged_df)

final_df
```

```python
final_df.to_excel('관광지_숙소 거리.xlsx')
```
