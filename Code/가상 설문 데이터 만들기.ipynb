```python
import pandas as pd
import numpy as np

# 설정값
num_samples = 1000
np.random.seed(0)  # 랜덤 시드 설정 (재현성을 위해)

# 1. 성별
genders = np.random.choice([1, 2], size=num_samples, p=[0.5, 0.5])

# 2. 나이
age_probabilities = [29.8, 34.8, 20.5, 14.9]  # 비율
age_probabilities_normalized = np.array(age_probabilities) / sum(age_probabilities)  # 확률로 정규화
ages = np.random.choice([1, 2, 3, 4], size=num_samples, p=age_probabilities_normalized)

# 3. 혼인 여부
marital_status = np.random.choice([1, 2], size=num_samples, p=[0.6, 0.4])

# 4. 방문 횟수
visit_counts = np.random.choice([1, 2, 3, 4, 5], size=num_samples, p=[0.4, 0.3, 0.15, 0.1, 0.05])

# 5. 워케이션 장소 선택 이유
# 비율에 따라 각 선택지의 개수 생성
reasons_counts = [
    int(num_samples * 15.656228157092272 / 100),
    int(num_samples * 11.789516087048174 / 100),
    int(num_samples * 33.88064885864273 / 100),
    int(num_samples * 30.418726674228207 / 100),
    int(num_samples * 4.381767300863091 / 100),
    int(num_samples * 2.7824077274500114 / 100),
    int(num_samples * 1.0907051946755153 / 100)
]

# 각 선택지 배열 생성
reasons = np.concatenate([
    np.full(reasons_counts[0], 1),
    np.full(reasons_counts[1], 2),
    np.full(reasons_counts[2], 3),
    np.full(reasons_counts[3], 4),
    np.full(reasons_counts[4], 5),
    np.full(reasons_counts[5], 6),
    np.full(reasons_counts[6], 7)
])

# 배열을 랜덤하게 섞어 데이터 무작위화
np.random.shuffle(reasons)

# 배열 길이 맞추기
if len(reasons) < num_samples:
    reasons = np.concatenate([reasons, np.full(num_samples - len(reasons), 1)])
elif len(reasons) > num_samples:
    reasons = reasons[:num_samples]

# 6. 동반 인원수
companions = np.random.choice([1, 2, 3, 4], size=num_samples, p=[0.4, 0.3, 0.2, 0.1])

# 7. 반려동물을 키우십니까
pets = np.random.choice([1, 2], size=num_samples, p=[0.29, 0.71])

# 데이터프레임 생성
df = pd.DataFrame({
    '성별': genders,
    '나이': ages,
    '혼인 여부': marital_status,
    '방문 횟수': visit_counts,
    '워케이션 장소 선택 이유': reasons,
    '동반 인원수': companions,
    '반려동물을 키우십니까': pets
})

df
```

```python
# 파일 경로 설정
file_path = '/content/drive/MyDrive/천안시 공모전/검색 건수 데이터/24.1-24.7 천안시 전체 검색 건수.csv'

# CSV 파일로 저장
df.to_csv(file_path, index=False, encoding='utf-8-sig')

print(f"가상 설문 데이터가 '{file_path}' 파일로 저장되었습니다.")
```
