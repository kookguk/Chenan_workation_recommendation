```python
df=pd.read_excel('천안시 관광지_콘텐츠 추천.xlsx')
df1=pd.read_excel('천안시 음식_콘텐츠 추천.xlsx')
df
```

```python
df = df.rename(columns={
    '도로명주소': '관광지주소',
    'Latitude': '관광지 Latitude',
    'Longitude': '관광지 Longitude',
    '중분류 카테고리': '관광지 중분류',
    '소분류 카테고리': '관광지 소분류',
    '검색건수': '관광지 검색건수',
    '유사도1' : '관광지 유사도1',
    '유사도2' : '관광지 유사도2',
    '유사도3' : '관광지 유사도3'
})

df = df[['관광지명', '관광지주소', '관광지 Latitude', '관광지 Longitude', '관광지 중분류', '관광지 소분류', '관광지 검색건수','유사 관광지명','관광지 유사도1','관광지 유사도2','관광지 유사도3']]
df
```

```python
df1 = df1.rename(columns={
    '관광지명': '식당명',
    '도로명주소': '식당주소',
    'Latitude': '식당 Latitude',
    'Longitude': '식당 Longitude',
    '중분류 카테고리': '식당 중분류',
    '소분류 카테고리': '식당 소분류',
    '검색건수': '식당 검색건수',
    '유사 관광지명' : '유사 식당명',
    '유사도1' : '식당 유사도1',
    '유사도2' : '식당 유사도2',
    '유사도3' : '식당 유사도3'
})

df1 = df1[['식당명', '식당주소', '식당 Latitude', '식당 Longitude', '식당 중분류', '식당 소분류', '식당 검색건수', '유사 식당명', '식당 유사도1','식당 유사도2','식당 유사도3']]
df1
```

```python
from geopy.distance import geodesic

def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).km

merged_df = []

for i, row in df.iterrows():
    관광지_lat = row['관광지 Latitude']
    관광지_lon = row['관광지 Longitude']
    
    df1['거리'] = df1.apply(lambda x: calculate_distance(관광지_lat, 관광지_lon, x['식당 Latitude'], x['식당 Longitude']), axis=1)
    
    nearby_attractions = df1[df1['거리'] <= 2].copy()
    
    if not nearby_attractions.empty:
        for _, attraction_row in nearby_attractions.iterrows():
            combined_row = row.to_dict()
            combined_row.update(attraction_row.to_dict())
            merged_df.append(combined_row)

final_df = pd.DataFrame(merged_df)

final_df
```

```python
final_df.to_excel('관광지_식당 거리.xlsx')
```
