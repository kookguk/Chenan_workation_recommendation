```python
df2=pd.read_excel('/content/drive/MyDrive/천안시 공모전/협업 필터링 사용 데이터/천안관광지_별점.xlsx')
df2
```

```python
# df2에서 점수 데이터 정리
df2 = df2[df2['점수'] != '점수 없음']  # '점수 없음' 제거
df2['점수'] = df2['점수'].astype(float)  # 점수를 float로 변환
tourist_mean_scores = df2['점수'].mean()  # 관광지 평균 점수

# 사용자 수와 관광지 수 정의
num_users = df1.shape[0]
num_spots = df2.shape[0]

# 개인별 점수 매기기
user_scores = []

for user_id in range(num_users):
    # 랜덤으로 0~5 사이의 점수 생성
    scores = np.random.rand(num_spots) * 5
    average_score = np.mean(scores)  # 생성된 점수의 평균

    # 각 관광지 평균 점수에 맞춰 점수 조정
    adjustment = tourist_mean_scores - average_score
    adjusted_scores = scores + adjustment

    # 점수가 0~5 범위로 제한
    adjusted_scores = np.clip(adjusted_scores, 0, 5)

    user_scores.append(adjusted_scores)

# 개인별 점수 DataFrame 생성
user_ratings = pd.DataFrame(user_scores, columns=df2['관광지'])
user_ratings['User'] = range(num_users)

# 결과 출력
print(user_ratings)

# 전체 평균 점수 확인
print("전체 평균 점수:", user_ratings.mean())
```

```python
# df2라는 DataFrame을 해당 경로에 '관광지 전처리.csv'라는 이름으로 저장
file_path = '/content/drive/MyDrive/천안시 공모전/협업 필터링 사용 데이터/관광지 전처리.csv'
df2.to_csv(file_path, index=False, encoding='utf-8-sig')

print(f"df2가 '{file_path}'에 저장되었습니다.")
```

```python
# df1 데이터와 user_ratings 결합하기
result_df = pd.concat([df1, user_ratings], axis=1)

# 결과 출력
print(result_df)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 1. 데이터 전처리
# 사용자 정보 데이터 df1
# 관광지 점수 데이터 user_ratings

# 2. 사용자 유사도 계산 (코사인 유사도)
user_features = df1[['성별', '나이', '혼인 여부', '방문 횟수', '워케이션 장소 선택 이유', '동반 인원수', '반려동물을 키우십니까']]
user_similarity = cosine_similarity(user_features)

# 3. 예상 점수 계산을 위한 함수 정의
def predict_ratings(user_similarity, user_ratings):
    user_mean = user_ratings.mean(axis=1).values.reshape(-1, 1)  # 1차원 배열을 2차원으로 변환
    ratings_diff = user_ratings.values - user_mean  # NumPy 배열 연산
    pred = user_mean + user_similarity.dot(ratings_diff) / np.abs(user_similarity).sum(axis=1).reshape(-1, 1)
    return pred

# 예상 점수 계산
predicted_ratings = predict_ratings(user_similarity, user_ratings)

# 4. 각 사용자에게 상위 30개 관광지 추천
def recommend_top_n_items(predicted_ratings, n=30):
    recommendations = np.argsort(predicted_ratings, axis=1)[:, -n:][:, ::-1]
    return recommendations

top_30_recommendations = recommend_top_n_items(predicted_ratings, n=30)

# 결과 출력
for user_id in range(len(top_30_recommendations)):
    print(f"사용자 {user_id}에게 추천하는 관광지: {top_30_recommendations[user_id]}")
```

```python
# 5. 추천 결과를 열별로 정리하여 DataFrame 생성
recommendation_data = []
for user_id, recommendations in enumerate(top_30_recommendations):
    recommendation_data.append([user_id + 1] + list(recommendations))  # user_id + 1로 User_ID 시작

columns = ['User_ID'] + [f'추천장소{i+1}' for i in range(30)]
recommendation_df = pd.DataFrame(recommendation_data, columns=columns)
```
